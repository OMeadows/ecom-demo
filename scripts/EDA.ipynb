```python
import pandas as pd
```

# Read in the files and join the dataframes together to create a olist_market_funnel_complete dataframe and a olist_orders_complete dataframe

```python
# Reading all the files
olist_customer = pd.read_csv('/kaggle/input/brazilian-ecommerce/olist_customers_dataset.csv')
olist_geolocation = pd.read_csv('/kaggle/input/brazilian-ecommerce/olist_geolocation_dataset.csv')
olist_orders = pd.read_csv('/kaggle/input/brazilian-ecommerce/olist_orders_dataset.csv')
olist_order_items = pd.read_csv('/kaggle/input/brazilian-ecommerce/olist_order_items_dataset.csv')
olist_order_payments = pd.read_csv('/kaggle/input/brazilian-ecommerce/olist_order_payments_dataset.csv')
olist_order_reviews = pd.read_csv('/kaggle/input/brazilian-ecommerce/olist_order_reviews_dataset.csv')
olist_products = pd.read_csv('/kaggle/input/brazilian-ecommerce/olist_products_dataset.csv')
olist_sellers = pd.read_csv('/kaggle/input/brazilian-ecommerce/olist_sellers_dataset.csv')
olist_product_category_name_translation = pd.read_csv('/kaggle/input/brazilian-ecommerce/product_category_name_translation.csv')
olist_closed_deals = pd.read_csv('/kaggle/input/marketing-funnel-olist/olist_closed_deals_dataset.csv')
olist_marketing_qualified_leads = pd.read_csv('/kaggle/input/marketing-funnel-olist/olist_marketing_qualified_leads_dataset.csv')
```

```python
olist_marketing_funnel = olist_marketing_qualified_leads.merge(olist_closed_deals, on = 'mql_id', how = 'left') ## joining the marketing funnels together
olist_marketing_funnel_sellers = olist_sellers.merge(olist_marketing_funnel, on= 'seller_id', how = 'left') ## can now see sellers location 
olist_marketing_funnel_sellers_and_items= olist_marketing_funnel_sellers.merge(olist_order_items, on = 'seller_id', how = 'left') ## can see what products the sellers sold 
olist_marketing_funnel_complete = pd.merge(olist_marketing_funnel_sellers_and_items, olist_geolocation,how = 'left', left_on = 'seller_zip_code_prefix', right_on = 'geolocation_zip_code_prefix') ##geolocation for seller 

olist_marketing_funnel_complete.info()
```

```python
olist_orders_complete = olist_orders.merge(olist_order_items, on = 'order_id', how = 'left') ## info on the items in specific orders 
olist_orders_complete = olist_orders_complete.merge(olist_order_payments, on = 'order_id', how = 'left') ## payments of those orders
olist_orders_complete= olist_orders_complete.merge(olist_order_reviews, on = 'order_id', how = 'left') ## reviews of the orders
```

```python
olist_product_category_name_translation.rename(columns = {'product_category_name': 'product_category_brazilian'}, inplace = True)
olist_product_category_name_translation.rename(columns = {'product_category_name_english': 'product_category_name'}, inplace = True)
olist_products = olist_products.merge(olist_product_category_name_translation, on = 'product_category_name', how = 'left')

olist_orders_complete = olist_orders_complete.merge(olist_products, on = 'product_id', how = 'left') ## specific product information for each order
```

```python
olist_orders_complete = olist_orders_complete.merge(olist_customer, on = 'customer_id', how = 'left') ## see customer information
olist_orders_complete = pd.merge(olist_orders_complete, olist_geolocation,how = 'left', left_on = 'customer_zip_code_prefix', right_on = 'geolocation_zip_code_prefix') ## where customers are buying from
```
